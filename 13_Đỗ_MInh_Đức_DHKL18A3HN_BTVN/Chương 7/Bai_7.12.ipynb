```python
import tkinter as tk

def tinh_bmi():
    can_nang = float(entry_weight.get())
    chieu_cao = float(entry_height.get()) / 100
    bmi = can_nang / (chieu_cao * chieu_cao)
    lbl_result.config(text=f"BMI = {bmi:.2f}")

root = tk.Tk()
root.title("BMI Calculator")

tk.Label(root, text="Cân nặng (kg):").pack()
entry_weight = tk.Entry(root)
entry_weight.pack()

tk.Label(root, text="Chiều cao (cm):").pack()
entry_height = tk.Entry(root)
entry_height.pack()

tk.Button(root, text="Tính BMI", command=tinh_bmi).pack(pady=5)
lbl_result = tk.Label(root, text="")
lbl_result.pack()

root.mainloop()
```
