```python
import tkinter as tk

def kiem_tra_du_lieu(*args):
    if entry_name.get() and entry_id.get() and entry_pass.get():
        btn_submit.pack(pady=10)
    else:
        btn_submit.pack_forget()


window = tk.Tk()
window.title("Thông tin sinh viên")
window.geometry("300x200")

tk.Label(window, text="Tên sinh viên:").pack()
entry_name = tk.Entry(window)
entry_name.pack()

tk.Label(window, text="ID sinh viên:").pack()
entry_id = tk.Entry(window)
entry_id.pack()

tk.Label(window, text="Mật khẩu:").pack()
entry_pass = tk.Entry(window, show="*")
entry_pass.pack()

btn_submit = tk.Button(window, text="Gửi")

entry_name.bind("<KeyRelease>", kiem_tra_du_lieu)
entry_id.bind("<KeyRelease>", kiem_tra_du_lieu)
entry_pass.bind("<KeyRelease>", kiem_tra_du_lieu)

window.mainloop()
```

```python
import tkinter as tk

def dao_nguoc():
    s = entry.get()
    kq = ""
    for i in range(len(s)-1, -1, -1):
        kq += s[i]
    lbl_result.config(text=kq)

root = tk.Tk()
root.title("Reverse Word")

tk.Label(root, text="Enter a word:").pack()
entry = tk.Entry(root, width=30)
entry.pack()

tk.Button(root, text="Validate", command=dao_nguoc).pack(pady=5)
lbl_result = tk.Label(root, text="")
lbl_result.pack()

root.mainloop()
```

```python
import tkinter as tk

def tinh_tong():
    n = int(entry.get())
    tong = 0
    for i in range(1, n+1):
        tong += i
    lbl.config(text=f"The sum is 1 + 2 + ... + {n} = {tong}")

root = tk.Tk()
root.title("Sum")

tk.Label(root, text="Enter value of integer N:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Validate", command=tinh_tong).pack(pady=5)
lbl = tk.Label(root, text="")
lbl.pack()

root.mainloop()
```

```python
import tkinter as tk

def liet_ke_uoc():
    n = int(entry.get())
    uoc = []
    for i in range(1, n+1):
        if n % i == 0:
            uoc.append(str(i))
    lbl.config(text="The divisors of N: " + "  ".join(uoc))

root = tk.Tk()
root.title("Divisors")

tk.Label(root, text="Enter the value of N:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Validate", command=liet_ke_uoc).pack(pady=5)
lbl = tk.Label(root, text="")
lbl.pack()

root.mainloop()
```

```python
import tkinter as tk
from math import gcd

def tinh():
    m = int(e1.get())
    n = int(e2.get())
    if choice.get() == "GCD":
        kq = gcd(m, n)
        lbl.config(text=f"GCD({m}, {n}) = {kq}")
    else:
        kq = m * n // gcd(m, n)
        lbl.config(text=f"LCM({m}, {n}) = {kq}")

root = tk.Tk()
root.title("GCD - LCM")

tk.Label(root, text="Enter value of m:").pack()
e1 = tk.Entry(root)
e1.pack()

tk.Label(root, text="Enter value of n:").pack()
e2 = tk.Entry(root)
e2.pack()

choice = tk.StringVar(value="GCD")
tk.OptionMenu(root, choice, "GCD", "LCM").pack()

tk.Button(root, text="Validate", command=tinh).pack(pady=5)
lbl = tk.Label(root, text="")
lbl.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import simpledialog, messagebox

def nhap_tuoi():
    tuoi = simpledialog.askinteger(
        "Nhập tuổi",
        "Vui lòng nhập tuổi của bạn:"
    )

    if tuoi is None:
        return
    elif tuoi <= 0:
        messagebox.showerror(
            "Thông báo",
            "Xin chào, sao lại tèn tèn quá vậy"
        )
    elif tuoi >= 18:
        messagebox.showinfo(
            "Thông báo",
            "Bạn đã trưởng thành"
        )
    else:
        messagebox.showwarning(
            "Thông báo",
            "Bạn còn nhỏ tuổi"
        )

root = tk.Tk()
root.title("Bài 7.10 - Kiểm tra tuổi")
root.geometry("300x150")

btn = tk.Button(root, text="Nhập tuổi", command=nhap_tuoi)
btn.pack(expand=True)

root.mainloop()
```

```python
import tkinter as tk

can = ["Canh", "Tân", "Nhâm", "Quý", "Giáp",
       "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
chi = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu",
       "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]

def doi_nam():
    year = int(entry.get())
    lbl.config(text=can[year % 10] + " " + chi[year % 12])

root = tk.Tk()
root.title("Âm lịch")

tk.Label(root, text="Nhập năm dương lịch:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Chuyển đổi", command=doi_nam).pack(pady=5)
lbl = tk.Label(root, text="")
lbl.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show_info(bai_tap):
    if bai_tap == "1.1":
        messagebox.showinfo("Lập trình hướng đối tượng", "Giải phương trình bậc 2")
    else:
        messagebox.showinfo("Thông báo", f"Bạn đã chọn bài tập {bai_tap}")

root = tk.Tk()
root.title("Hệ thống bài tập nâng cao Python")
root.geometry("500x300")

menubar = tk.Menu(root)

for i in range(1, 8):
    chuong_menu = tk.Menu(menubar, tearoff=0)
    for j in range(1, 10):
        bai = f"{i}.{j}"
        chuong_menu.add_command(label=bai, command=lambda b=bai: show_info(b))
    
    chuong_menu.add_separator()
    chuong_menu.add_command(label="Exit", command=root.quit)
    menubar.add_cascade(label=f"Chương {i}", menu=chuong_menu)

root.config(menu=menubar)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def tinh_tien():
    try:
        km = float(entry_km.get())
        phut_cho = int(entry_cho.get())
        loai_xe = var_xe.get()
        
        if loai_xe == 4:
            gia_duoi_30 = 15300
            gia_tren_30 = 12100
        else: # Xe 7 chỗ
            gia_duoi_30 = 16100
            gia_tren_30 = 13800
            
        if km <= 0.8:
            tien_xe = 12000
        elif km <= 30:
            tien_xe = 12000 + (km - 0.8) * gia_duoi_30
        else:
            tien_xe = 12000 + (30 - 0.8) * gia_duoi_30 + (km - 30) * gia_tren_30
            
        tien_cho = max(0, phut_cho - 5) * 750
        tong_tien = tien_xe + tien_cho
        
        lbl_kq.config(text=f"Tổng cước: {tong_tien:,.0f} VNĐ")
        
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập số liệu hợp lệ!")

app = tk.Tk()
app.title("Tính cước Taxi")
app.geometry("350x400")

tk.Label(app, text="LOẠI XE", font=('Arial', 10, 'bold')).pack(pady=5)
var_xe = tk.IntVar(value=4)
tk.Radiobutton(app, text="Xe 4 chỗ", variable=var_xe, value=4).pack()
tk.Radiobutton(app, text="Xe 7 chỗ", variable=var_xe, value=7).pack()

tk.Label(app, text="Số km di chuyển:").pack(pady=5)
entry_km = tk.Entry(app)
entry_km.pack()

tk.Label(app, text="Số phút chờ:").pack(pady=5)
entry_cho = tk.Entry(app)
entry_cho.pack()

tk.Button(app, text="TÍNH CƯỚC", command=tinh_tien, bg="blue", fg="white").pack(pady=20)
lbl_kq = tk.Label(app, text="Tổng cước: 0 VNĐ", font=('Arial', 12, 'bold'), fg="red")
lbl_kq.pack()

app.mainloop()
```
