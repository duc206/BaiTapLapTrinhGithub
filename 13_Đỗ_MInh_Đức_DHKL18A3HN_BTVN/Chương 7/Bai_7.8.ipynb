```python
import tkinter as tk

def liet_ke_uoc():
    n = int(entry.get())
    uoc = []
    for i in range(1, n+1):
        if n % i == 0:
            uoc.append(str(i))
    lbl.config(text="The divisors of N: " + "  ".join(uoc))

root = tk.Tk()
root.title("Divisors")

tk.Label(root, text="Enter the value of N:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Validate", command=liet_ke_uoc).pack(pady=5)
lbl = tk.Label(root, text="")
lbl.pack()

root.mainloop()
```
