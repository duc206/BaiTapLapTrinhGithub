```python
import tkinter as tk

def dao_nguoc():
    s = entry.get()
    kq = ""
    for i in range(len(s)-1, -1, -1):
        kq += s[i]
    lbl_result.config(text=kq)

root = tk.Tk()
root.title("Reverse Word")

tk.Label(root, text="Enter a word:").pack()
entry = tk.Entry(root, width=30)
entry.pack()

tk.Button(root, text="Validate", command=dao_nguoc).pack(pady=5)
lbl_result = tk.Label(root, text="")
lbl_result.pack()

root.mainloop()
```
