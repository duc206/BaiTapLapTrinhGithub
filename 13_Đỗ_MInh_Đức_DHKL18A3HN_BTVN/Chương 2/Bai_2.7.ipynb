```python
import json

class Student:
    def __init__(self, name, age, scores):
        self.name = name
        self.age = age
        self.scores = scores

    def __str__(self):
        return f"{self.name} - {self.age} - {self.scores}"


json_data = '''
{
  "name": "Nguyen Van A",
  "age": 22,
  "scores": [9, 9,8]
}
'''

data = json.loads(json_data)
student = Student(**data)

print(student)
```
