```python
import json


data = {
    "name": "Nguyen Thi B",
    "age": 24,
    "scores": [9, 10, 7],
    "passed": True
}


json_str = json.dumps(data, ensure_ascii=False)

print("Chuỗi JSON:")
print(json_str)

print("\nCác giá trị trong đối tượng:")
for value in data.values():
    print(value)
```

```python
import json

class Student:
    def __init__(self, name, age, scores):
        self.name = name
        self.age = age
        self.scores = scores
```

```python
sv = Student("Nguyen Van B", 25, [7, 8, 9])

json_str = json.dumps(sv.__dict__, ensure_ascii=False)

print("Chuỗi JSON:")
print(json_str)

print("\nCác giá trị của đối tượng:")
for value in sv.__dict__.values():
    print(value)
```

```python
obj = json.loads(json_str)

print("\nCác giá trị trong JSON:")
for v in obj.values():
    print(v)
```
