```python
company_info = {
        "ten_cong_ty": "Công ty TNHH Đất Việt",
        "dia_chi": "abc Giải Phóng - Hà Nội",
        "nhan_vien": [
            {"ten": "Nguyen Van A", "don_vi": "Đơn vị A1"},
            {"ten": "Tran Thi B", "don_vi": "Đơn vị A2"},
            {"ten": "Le Van C", "don_vi": "Đơn vị A1"},
            {"ten": "Pham Thi D", "don_vi": "Đơn vị A3"},
            {"ten": "Hoang Van E", "don_vi": "Đơn vị A2"},
        ]
    }

nhan_vien = company_info["nhan_vien"]
tong_nv = len(nhan_vien)

don_vi_count = {}
for nv in nhan_vien:
    don_vi = nv["don_vi"]
    don_vi_count[don_vi] = don_vi_count.get(don_vi, 0) + 1

print(f"Tên công ty: {company_info['ten_cong_ty']}")
print(f"Địa chỉ: {company_info['dia_chi']}")
print("..--Thống kê nhân viên theo đơn vị---")
for idx, (dv, count) in enumerate(don_vi_count.items(), 1):
    ty_le = count / tong_nv * 100
    print(f"{idx}. Tên đơn vị: {dv}")
    print(f"   Số nhân viên: {count}")
    print(f"   Tỷ lệ: {ty_le:.2f}%")
```

```python
import datetime
giao_dich = []
while True:
    loai = input("Nhập loại giao dịch (tien/vang) hoặc 'exit' để kết thúc: ").strip()
    if loai.lower() == "exit":
        break
    so_tien = float(input("Nhập số tiền/gold: "))
    giao_dich.append({"loai": loai, "so_tien": so_tien})

ghi_file = input("Bạn có muốn ghi dữ liệu vào file không? (1: Có, 0: Không): ").strip()
if ghi_file == "1":
    ngay_gio = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
    ten_file = f"{ngay_gio}.json"
    with open(ten_file, "w", encoding="utf-8") as f:
        json.dump(giao_dich, f, indent=4, ensure_ascii=False)
    print(f"Đã ghi dữ liệu vào file {ten_file}")
else:
    print("Không ghi dữ liệu.")
```

```python
import requests
url = "https://jsonplaceholder.typicode.com/posts"
try:
    response = requests.get(url)
    posts = response.json()
    print(f"Tổng số bài post: {len(posts)}\n")
    for post in posts[:5]:  
        print(f"UserID: {post['userId']}")
        print(f"ID: {post['id']}")
        print(f"Title: {post['title']}")
        print(f"Body: {post['body']}\n")
except Exception as e:
    print("Lỗi khi gọi API:", e)
```

```python
url = "https://jsonplaceholder.typicode.com/comments?postId=1"
    
try:
    response = requests.get(url)
    response.raise_for_status()  # kiểm tra lỗi HTTP
    posts = response.json()      # chuyển JSON về Python list
    for i, post in enumerate(posts[:3], 1):
        print(f"\nBài post số {i}:")
        print(f"ID: {post['id']}")
        print(f"Post ID: {post['postId']}")
        print(f"Tên: {post['name']}")
        print(f"Email: {post['email']}")
        print(f"Nội dung: {post['body']}")
            
except requests.exceptions.RequestException as e:
    print("Lỗi khi gọi API:", e)
```
