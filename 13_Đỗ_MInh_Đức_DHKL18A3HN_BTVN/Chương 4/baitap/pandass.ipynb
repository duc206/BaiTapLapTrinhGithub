```python
stocks1 = pd.read_csv("stocks1.csv")
print(stocks1.head())
print(stocks1.tail())
print(stocks1.dtypes)
print(stocks1.info())
stocks2 = pd.read_csv("stocks2.csv")
print(stocks2.head())
print(stocks2.tail())
print(stocks2.dtypes)
print(stocks2.info())
companies = pd.read_csv("companies.csv")
print(companies.head())
print(companies.dtypes)
print(companies.info())
print(stocks1.isnull().sum())
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))
stocks = pd.concat([stocks1, stocks2], ignore_index=True)
print(stocks.tail(15))
stocks_companies = pd.merge(stocks, companies, left_on='symbol', right_on='name', how='left')
print(stocks_companies.head())
avg_price_volume = stocks_companies.groupby('symbol')[['open','high','low','close','volume']].mean()
print(avg_price_volume)
close_stats = stocks_companies.groupby('symbol')['close'].agg(['mean','max','min'])
print(close_stats)
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print(stocks_companies['parsed_time'].dtype)
print(stocks_companies.head())
stocks_companies['result'] = stocks_companies.apply(lambda row: 'up' if row['close'] > row['open'] else 'down', axis=1)
print(stocks_companies.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

movie = pd.read_csv("movies.csv")

print(type(movie))
print(movie.shape)
print(movie.columns)
print(movie.head())
print(movie.tail())

print(movie.isna().sum())
movie["genres"] = movie["genres"].fillna("Unknown")

year_series = movie["title"].str.extract(r'\((\d{4})\)') 
movie["year"] = pd.to_numeric(year_series.iloc[:, 0], errors="coerce")

movie_expanded = movie.copy()
movie_expanded["genres"] = movie_expanded["genres"].str.split("|")
movie_expanded = movie_expanded.explode("genres")

filtered_action = movie_expanded[movie_expanded["genres"].str.contains("Action", na=False)]
print(filtered_action.head())

movies_per_year = movie["year"].dropna().astype(int).value_counts().sort_index()
print(movies_per_year)

if not movies_per_year.empty:
    plt.figure(figsize=(10,5))
    movies_per_year.plot(kind="line")
    plt.title("Số phim theo từng năm")
    plt.xlabel("Năm")
    plt.ylabel("Số lượng phim")
    plt.tight_layout()
    plt.show()
else:
    print("Không có dữ liệu năm hợp lệ để vẽ biểu đồ.")
```
