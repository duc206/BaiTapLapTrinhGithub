```python
import threading

def in_so_chan():
    print("Các số chẵn:")
    for i in range(30, 51):
        if i % 2 == 0:
            print(i)

def in_so_le():
    print("Các số lẻ:")
    for i in range(30, 51):
        if i % 2 != 0:
            print(i)


luong1 = threading.Thread(target=in_so_chan, name="Luong-So-Chan")
luong2 = threading.Thread(target=in_so_le, name="Luong-So-Le")

luong1.start()
luong2.start()

luong1.join()
luong2.join()

print("Chương trình kết thúc")
```
