```python
import threading
import random

max_cac_thread = []

def tim_max_list_con(list_con):
    m = max(list_con)
    max_cac_thread.append(m)
    print(f"{threading.current_thread().name}: {list_con} -> Max = {m}")

n = int(input("Nhập số phần tử: "))
lst = [random.randint(0, 100) for _ in range(n)]
print("List:", lst)

so_thread = int(input("Nhập số thread: "))

kich_thuoc = n // so_thread
threads = []

for i in range(so_thread):
    start = i * kich_thuoc
    end = n if i == so_thread - 1 else (i + 1) * kich_thuoc
    list_con = lst[start:end]
    t = threading.Thread(
        target=tim_max_list_con,
        args=(list_con,),
        name=f"Thread-{i+1}"
    )
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Giá trị lớn nhất trong list:", max(max_cac_thread))
```
