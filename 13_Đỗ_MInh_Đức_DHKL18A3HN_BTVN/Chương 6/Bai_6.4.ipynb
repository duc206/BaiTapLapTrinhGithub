```python
import threading

ket_qua = []

def tinh_doan(start, end):
    tich = 1
    for i in range(start, end + 1):
        tich *= i
    ket_qua.append(tich)

n = 11

t1 = threading.Thread(target=tinh_doan, args=(1, n // 2))
t2 = threading.Thread(target=tinh_doan, args=(n // 2 + 1, n))

t1.start()
t2.start()

t1.join()
t2.join()

giai_thua = 1
for k in ket_qua:
    giai_thua *= k

print(f"{n}! =", giai_thua)
```
