```python
import threading
import time

def chay_web(ten):
    print(f"Starting {ten}")
    for _ in range(3):
        print(f"{ten}: {time.ctime()}")
        time.sleep(1)
    print(f"Exiting {ten}")

print("Starting Main Thread")

t1 = threading.Thread(target=chay_web, args=("Google",))
t2 = threading.Thread(target=chay_web, args=("Yahoo",))
t3 = threading.Thread(target=chay_web, args=("Facebook",))

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```
