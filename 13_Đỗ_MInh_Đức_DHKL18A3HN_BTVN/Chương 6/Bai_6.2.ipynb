```python
import threading
import requests

def tai_tep(url, ten_file):
    print(f"Bắt đầu tải: {ten_file}")
    response = requests.get(url)
    with open(ten_file, "wb") as f:
        f.write(response.content)
    print(f"Tải xong: {ten_file}")

urls = [
    ("https://example.com/file1.jpg", "file1.jpg"),
    ("https://example.com/file2.jpg", "file2.jpg"),
    ("https://example.com/file3.jpg", "file3.jpg")
]

threads = []

# Tạo và chạy các luồng
for url, ten_file in urls:
    t = threading.Thread(target=tai_tep, args=(url, ten_file))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Đã tải xong tất cả các tệp")
```
