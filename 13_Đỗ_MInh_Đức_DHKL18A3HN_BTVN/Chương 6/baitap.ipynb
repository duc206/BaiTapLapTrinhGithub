```python
print("Chương trinhg python chạy nhiều luồng(Mulyo threads)")
import threading
import time
def worker(num):
    print(f'Worker {num} bắt đầu')
    time.sleep(2)
    print(f'Worker {num} kết thúc')
threads = []
for i in range(5):
    t = threading.Thread(target=worker, args=(i,))
    threads.append(t)
    t.start()
for t in threads:
    t.join()
print("Tất cả các luồng đã hoàn thành")
```

```python
import threading
import requests
urls = [
    "https://example.com/file1.txt",
    "https://example.com/file2.txt"
]

def download(url):
    ten_file = url.split("/")[-1]
    response = requests.get(url)

    with open(ten_file, "wb") as f:
        f.write(response.content)

    print("Đã tải xong:", ten_file)

threads = []

for url in urls:
    t = threading.Thread(target=download, args=(url,))
    threads.append(t)
    t.start()
for t in threads:
    t.join()

print(" Tải xong tất cả file")
```

```python
print("Đọc số chẵn và lẻ từ 30 đến 50")
def read_even():
    for i in range(30, 51):
        if i % 2 == 0:
            print(f'Số chẵn: {i}')
            time.sleep(0.1)
def read_odd():
    for i in range(30, 51):
        if i % 2 != 0:
            print(f'Số lẻ: {i}')
            time.sleep(0.1)
even_thread = threading.Thread(target=read_even)
odd_thread = threading.Thread(target=read_odd)  
even_thread.start()
odd_thread.start()  
even_thread.join()
odd_thread.join()
print("Hoàn thành đọc số chẵn và lẻ")   
```

```python
import threading

n = int(input("Nhập một số nguyên dương n: "))
so_luong_luong = int(input("Nhập số lượng luồng: "))

doan = n // so_luong_luong

def giai_thua(start, end, index, results):
    result = 1
    for i in range(start, end + 1):
        result *= i
    results[index] = result

threads = []
results = [1] * so_luong_luong

for i in range(so_luong_luong):
    start = i * doan + 1
    end = (i + 1) * doan if i != so_luong_luong - 1 else n

    t = threading.Thread(
        target=giai_thua,
        args=(start, end, i, results)
    )
    threads.append(t)
    t.start()

for t in threads:
    t.join()

final_result = 1
for res in results:
    final_result *= res

print(f"Giai thừa của {n} là: {final_result}")
print("Chương trình Python sử dụng nhiều luồng")
```

```python
print("Thực hiện yêu cầu HTTP đồng thời bằng các luồng" )
import threading
import requests
urls = [
    "https://httpbin.org/get",
    "https://httpbin.org/delay/2",
    "https://httpbin.org/uuid"
]

def send_request(url):
    try:
        response = requests.get(url)
        print(f"URL: {url}")
        print("Status:", response.status_code)
    except Exception as e:
        print(f"Lỗi khi gửi request tới {url}:", e)

threads = []
for url in urls:
    t = threading.Thread(target=send_request, args=(url,))
    threads.append(t)
    t.start()
for t in threads:
    t.join()

print(" Đã thực hiện xong tất cả yêu cầu HTTP đồng thời")
```

```python
import threading

N = int(input("Nhập ngưỡng N: "))
def in_so_chan():
    for i in range(2, N + 1, 2):
        print("Số chẵn:", i)
def in_so_le():
    for i in range(1, N + 1, 2):
        print("Số lẻ :", i)
t1 = threading.Thread(target=in_so_chan)
t2 = threading.Thread(target=in_so_le)
t1.start()
t2.start()
t1.join()
t2.join()
print(" In xong các số")
```

```python
import threading
import time

def web_thread(name, delay, times):
    print(f"Starting {name}")
    for _ in range(times):
        time.sleep(delay)
        print(f"{name}: Web {time.ctime()}")
    print(f"Exiting {name}")

def main():
    print("Starting Main Thread")

    t1 = threading.Thread(target=web_thread, args=("Google", 1, 5))
    t2 = threading.Thread(target=web_thread, args=("Yahoo", 2, 4))
    t3 = threading.Thread(target=web_thread, args=("Facebook", 3, 3))

    t1.start()
    t2.start()
    t3.start()

    t1.join()
    t2.join()
    t3.join()

    print("Exiting Main Thread")

if __name__ == "__main__":
    main()
```

```python
import threading
import random

def sum_sublist(sublist, index, results):
    print(f"Thread {index + 1}: ", end="")
    for x in sublist:
        print(x, end="; ")
    print()
    results[index] = sum(sublist)

def main():
    n = int(input("Nhập số phần tử: "))
    num_threads = int(input("Nhập số thread: "))

    data = [random.randint(0, 10) for _ in range(n)]
    print("List:", data)

    chunk_size = n // num_threads
    threads = []
    results = [0] * num_threads

    for i in range(num_threads):
        start = i * chunk_size
        end = n if i == num_threads - 1 else (i + 1) * chunk_size
        sublist = data[start:end]

        t = threading.Thread(
            target=sum_sublist,
            args=(sublist, i, results)
        )
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    print("Tổng list =", sum(results))

if __name__ == "__main__":
    main()
```

```python
import threading
import random

def max_sublist(sublist, index, results):
    print(f"Thread {index + 1}: {sublist}")
    results[index] = max(sublist)

def main():
    n = int(input("Nhập số phần tử: "))
    num_threads = int(input("Nhập số thread: "))

    data = [random.randint(0, 100) for _ in range(n)]
    print("List:", data)

    chunk_size = n // num_threads
    threads = []
    results = [0] * num_threads

    for i in range(num_threads):
        start = i * chunk_size
        end = n if i == num_threads - 1 else (i + 1) * chunk_size
        sublist = data[start:end]

        t = threading.Thread(
            target=max_sublist,
            args=(sublist, i, results)
        )
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    print("Giá trị lớn nhất trong list =", max(results))

if __name__ == "__main__":
    main()
```
