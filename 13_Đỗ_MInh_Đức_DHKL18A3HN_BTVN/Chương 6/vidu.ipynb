```python
import threading
def print_cube(num):
# function to print cube of given num
    print("Cube: {}".format(num * num * num))
def print_square(num):
# function to print square of given num
    print("Square: {}".format(num * num))
if __name__ == "__main__":
# creating thread
    t1 = threading.Thread(target=print_square, args=(10,))
    t2 = threading.Thread(target=print_cube, args=(10,))
    # start the threads
    t1.start()
    t2.start()
    # wait until both threads finish
    t1.join()
    t2.join()
    # continue with the main loop
    for i in range(5):
        print("Main loop: {}".format(i))
    print("Done!")
```

```python
import threading
import os
import time
if __name__ == "__main__":
    print("ID của tiến trình chính: {}".format(os.getpid()))
    print("Tên của luồng chính: {}".format(threading.current_thread().name))
    # Kiểm tra thời gian thực thi không dùng luồng
    start_time = time.time()
    print("\n--- Thực thi tuần tự ---")
    task1()
    task2()
    sequential_time = time.time() - start_time
    print("Thời gian thực thi tuần tự: {:.2f} giây".format(sequential_time))
    # Kiểm tra thời gian thực thi dùng luồng
    start_time = time.time()
    print("\n--- Thực thi với luồng ---")
    t1 = threading.Thread(target=task1, name='t1')
    t2 = threading.Thread(target=task2, name='t2')
    t1.start()
    t2.start()
    t1.join()
    t2.join()
    threaded_time = time.time() - start_time
    print("Thời gian thực thi với luồng: {:.2f} giây".format(threaded_time))
```

```python
import pandas as pd
import time

files = ['sales_feb.csv', 'sales_mar.csv', 'sales_jan.csv']

def read_files():
    start_time = time.time()
    
    for file in files:
        df = pd.read_csv(file)
        print(f"Đã đọc xong {file} với {len(df)} dòng.")
    
    end_time = time.time()
    print(f"Tổng thời gian đọc file: {end_time - start_time:.2f} giây")

read_files()

     
```

```python
import pandas as pd
import threading
import time

files = ['sales_feb.csv', 'sales_mar.csv', 'sales_apr.csv']

result = []

def load_2(filename):
    time.sleep(2)
    df = pd.read_csv(filename)
    result.append((filename, len(df)))
    print(f"Đọc xong {filename}: {len(df)} dòng")

def read():
    start_time = time.time()
    threads = []
    for file in files:
        t = threading.Thread(target=load_2, args=(file,))
        threads.append(t)
        t.start()
    for t in threads:
        t.join()
    end_time = time.time()
    print(f"Tổng thời gian đọc file với luồng: {end_time - start_time:.2f} giây")
    print("Kết quả đọc file:", result)

read()
```

```python
from concurrent.futures import ThreadPoolExecutor
import pandas as pd
import time
def load_csv(filename):
    print(f"Đang đọc {filename}")
    time.sleep(0.5) 
    df = pd.read_csv(filename)
    return (filename, len(df))
files = ['sales_jan.csv', 'sales_feb.csv', 'sales_mar.csv', ]
with ThreadPoolExecutor(max_workers=3) as executor:
    result = executor.map(load_csv, files)
for filename, row_count in result:
    print(f"Đã đọc xong {filename} với {row_count} dòng.")
```

```python
import threading
x = 0
def increment():
    global x
    tap = x
    time.sleep(0.00001)
    x = tap + 1
def thread_task():
    for _ in range(100000):
        increment()
def main_task():
    global x
    x = 0
    t1 = threading.Thread(target=thread_task)
    t2 = threading.Thread(target=thread_task)
    t1.start()
    t2.start()
    t1.join()
    t2.join()
for i in range(10):
    main_task()
    print("Lần lặp {0}: x = {1}".format(i, x))
```

```python
balance = 100
def add_money():
    global balance
    lap = balance
    time.sleep(0.001)
    balance = lap + 10
thread = [threading.Thread(target=add_money) for _ in range(2)]
for i in thread:
    i.start()
for i in thread:
    i.join()
print("Số dư cuối cùng:", balance)
    
```

```python
lock = threading.Lock()
balance = 100
def add_money():
    global balance
    with lock:
        lap = balance
        time.sleep(0.001)
        balance = lap + 10
thread = [threading.Thread(target=add_money) for _ in range(2)]
for i in thread:
    i.start()
for i in thread:
    i.join()
print("Số dư cuối cùng:", balance)
    
```

```python
lock = threading.Lock()
balance = 100
def add_money():
    global balance
    lock.acquire()
    try:
        lap = balance
        time.sleep(0.001)
        balance = lap + 10
    finally:
        lock.release()
thread = [threading.Thread(target=add_money) for _ in range(2)]
for i in thread:
    i.start()
for i in thread:
    i.join()
print("Số dư cuối cùng:", balance)
    
```

```python
sen = threading.Semaphore(3)
def access(id):
    sen.acquire()
    print(f"Thread {id} đang truy cập tài nguyên.")
    time.sleep(1)
    print(f"Thread {id} đã rời khỏi tài nguyên.")
    sen.release()
for i in range(6):
    threading.Thread(target=access, args=(i,)).start()
```
