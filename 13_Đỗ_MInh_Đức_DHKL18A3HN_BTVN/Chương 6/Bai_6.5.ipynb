```python
import threading
import requests

def gui_yeu_cau(url):
    print(f"Luồng {threading.current_thread().name} đang gửi yêu cầu tới {url}")
    response = requests.get(url)
    print(f"Luồng {threading.current_thread().name} nhận mã trạng thái: {response.status_code}")

urls = [
    "https://httpbin.org/get",
    "https://jsonplaceholder.typicode.com/posts/1",
    "https://jsonplaceholder.typicode.com/users/1",
    "https://jsonplaceholder.typicode.com/todos/1"
]

threads = []

for i, url in enumerate(urls):
    t = threading.Thread(target=gui_yeu_cau, args=(url,), name=f"Luong-{i+1}")
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Đã hoàn thành tất cả các yêu cầu HTTP")
```
