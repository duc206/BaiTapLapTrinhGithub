```python
import threading

N = 20

def ghi_so_chan():
    print("Luồng số chẵn:")
    for i in range(0, N + 1):
        if i % 2 == 0:
            print(i)

def ghi_so_le():
    print("Luồng số lẻ:")
    for i in range(0, N + 1):
        if i % 2 != 0:
            print(i)

t1 = threading.Thread(target=ghi_so_chan, name="Luong-Chan")
t2 = threading.Thread(target=ghi_so_le, name="Luong-Le")

t1.start()
t2.start()

t1.join()
t2.join()

print("Chương trình kết thúc")
```
