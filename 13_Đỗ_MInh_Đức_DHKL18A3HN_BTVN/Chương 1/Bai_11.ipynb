```python
class Tam_giac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def chu_vi(self):
        return self.a + self.b + self.c
```

```python
class Tam_giac_can(Tam_giac):
    def __init__(self, a, c):
        super().__init__(a, a, c)
```

```python
import math

class Tam_giac_vuong(Tam_giac):
    def __init__(self, a, b):
        c = math.sqrt(a*a + b*b)
        super().__init__(a, b, c)
```

```python
class Tam_giac_deu(Tam_giac_can):
    def __init__(self, a):
        super().__init__(a, a)
```

```python
def main_bai11():
    tg = Tam_giac(3, 4, 5)
    print("Chu vi tam giác thường:", tg.perimeter())

    tg_can = Tam_giac_can(5, 6)
    print("Chu vi tam giác cân:", tg_can.perimeter())

    tg_vuong = Tam_giac_vuong(3, 4)
    print("Chu vi tam giác vuông:", tg_vuong.perimeter())

    tg_deu = Tam_giac_deu(4)
    print("Chu vi tam giác đều:", tg_deu.perimeter())


main_bai11()
```
