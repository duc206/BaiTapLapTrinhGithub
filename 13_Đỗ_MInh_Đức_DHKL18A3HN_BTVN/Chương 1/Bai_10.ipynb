```python
class Diem:
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

```python
import math

class Ellipse(Diem):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a
        self.b = b

    def area(self):
        return math.pi * self.a * self.b
```

```python
class Circle(Ellipse):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
```

```python
def main_bai10():
    x = float(input("Nhập tọa độ x tâm: "))
    y = float(input("Nhập tọa độ y tâm: "))
    a = float(input("Nhập bán trục lớn a: "))
    b = float(input("Nhập bán trục nhỏ b: "))

    elip = Ellipse(x, y, a, b)
    print("Diện tích elip:", elip.area())

    r = float(input("\nNhập bán kính đường tròn: "))
    circle = Circle(x, y, r)
    print("Diện tích đường tròn:", circle.area())


main_bai10()
```
