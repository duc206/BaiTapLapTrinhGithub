```python
#bt1
class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.dai = dai
        self.rong = rong

    def nhap(self):
        self.dai = float(input("Nhập chiều dài: "))
        self.rong = float(input("Nhập chiều rộng: "))

    def tinh_chu_vi(self):
        return 2 * (self.dai + self.rong)

    def tinh_dien_tich(self):
        return self.dai * self.rong

    def hien_thi(self):
        print("----- Thông tin Hình Chữ Nhật -----")
        print(f"Chiều dài: {self.dai}")
        print(f"Chiều rộng: {self.rong}")
        print(f"Chu vi: {self.tinh_chu_vi()}")
        print(f"Diện tích: {self.tinh_dien_tich()}")
```

```python
#bt2
class ThiSinh:
    def __init__(self, ho_ten="", toan=0, ly=0, hoa=0):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def nhap(self):
        self.ho_ten = input("Nhập họ tên thí sinh: ")
        self.toan = float(input("Nhập điểm Toán: "))
        self.ly = float(input("Nhập điểm Lý: "))
        self.hoa = float(input("Nhập điểm Hóa: "))

    def hien_thi(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}")
        print(f"Tổng điểm: {self.tinh_tong()}")

    def tinh_tong(self):
        return self.toan + self.ly + self.hoa

ds_thi_sinh = []

n = int(input("Nhập số lượng thí sinh: "))
for i in range(n):
    print(f"\n--- Nhập thông tin thí sinh thứ {i+1} ---")
    ts = ThiSinh()
    ts.nhap()
    ds_thi_sinh.append(ts)

diem_chuan = 20

ds_trung_tuyen = [ts for ts in ds_thi_sinh if ts.tinh_tong() >= diem_chuan]

ds_trung_tuyen.sort(key=lambda ts: ts.tinh_tong(), reverse=True)

print("\n===== Danh sách thí sinh trúng tuyển =====")
for ts in ds_trung_tuyen:
    ts.hien_thi()
    print("-----------------------")
```

```python
#bt3
from math import gcd

class PhanSo:
    def __init__(self, tu=0, mau=1):
        if mau == 0:
            raise ValueError("Mẫu số không được bằng 0")
        self.tu = tu
        self.mau = mau
        self.rut_gon()

    def hop_le(self):
        return self.mau != 0

    def rut_gon(self):
        ucln = gcd(int(self.tu), int(self.mau))
        self.tu //= ucln
        self.mau //= ucln

    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        if self.mau == 0:
            raise ValueError("Mẫu số phải khác 0")
        self.rut_gon()

    def hien_thi(self):
        print(f"{self.tu}/{self.mau}")
```

```python
#bt4-6
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.data = []
    
    def push(self, value: float):
        if self.is_full():
            print("Ngăn xếp đầy!")
        else:
            self.data.append(value)
    
    def pop(self):
        if self.is_empty():
            print("Ngăn xếp rỗng!")
            return None
        return self.data.pop()
    
    def is_empty(self):
        return len(self.data) == 0
    
    def is_full(self):
        return len(self.data) == self.capacity

    def count(self):
        return len(self.data)

    def print(self):
        print("Nội dung Stack:", self.data)

```

```python
#bt7
import datetime

class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year
    
    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")
    
    def next(self):
        current = datetime.date(self.year, self.month, self.day)
        tomorrow = current + datetime.timedelta(days=1)
        return Date(tomorrow.day, tomorrow.month, tomorrow.year)
```

```python
#bt8
class Employee:
    def __init__(self, ho_ten, ngay_sinh: Date, ngay_vao_cty: Date):
        self.ho_ten = ho_ten
        self.ngay_sinh = ngay_sinh
        self.ngay_vao_cty = ngay_vao_cty

    def display(self):
        print("Họ tên:", self.ho_ten)
        print("Ngày sinh:", end=" "); self.ngay_sinh.display()
        print("Ngày vào công ty:", end=" "); self.ngay_vao_cty.display()

ds_nv = []
ds_nv.append(Employee("Nguyễn Văn A", Date(1,1,2000), Date(10,5,2022)))
ds_nv.append(Employee("Trần Thị B", Date(15,3,1999), Date(12,6,2023)))

print("\nDanh sách nhân viên:")
for nv in ds_nv:
    nv.display()
    print("-----------------")
```

```python
#bt9
class DaGiac:
    def __init__(self, canh=[]):
        self.canh = canh

    def chu_vi(self):
        return sum(self.canh)


class HinhBinhHanh(DaGiac):
    def __init__(self, day, cao, canh_ben):
        super().__init__([day, canh_ben, day, canh_ben])
        self.day = day
        self.cao = cao
    
    def dien_tich(self):
        return self.day * self.cao


class HinhChuNhat(HinhBinhHanh):
    def __init__(self, dai, rong):
        super().__init__(dai, rong, rong)
        self.dai = dai
        self.rong = rong

    def dien_tich(self):
        return self.dai * self.rong


class HinhVuong(HinhChuNhat):
    def __init__(self, canh):
        super().__init__(canh, canh)
        self.canh = canh
```

```python
#bt10
import math

class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

class Elip(Diem):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a  
        self.b = b  

    def dien_tich(self):
        return math.pi * self.a * self.b

class DuongTron(Elip):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
        self.r = r
```

```python
#bt11
class TamGiac:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c

    def chu_vi(self):
        return self.a + self.b + self.c

    def dien_tich(self):
        p = self.chu_vi() / 2
        return math.sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))


class TamGiacVuong(TamGiac):
    def la_vuong(self):
        sides = sorted([self.a, self.b, self.c])
        return math.isclose(sides[0]**2 + sides[1]**2, sides[2]**2)


class TamGiacCan(TamGiac):
    def la_can(self):
        return self.a == self.b or self.a == self.c or self.b == self.c


class TamGiacDeu(TamGiacCan):
    def __init__(self, a):
        super().__init__(a, a, a)

    def dien_tich(self):
        return (math.sqrt(3)/4) * self.a**2
```
