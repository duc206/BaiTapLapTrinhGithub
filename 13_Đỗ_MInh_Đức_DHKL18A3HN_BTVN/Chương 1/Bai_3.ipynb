```python
from math import gcd

class PhanSo:
    def __init__(self, tu=0, mau=1):
        if mau == 0:
            raise ValueError("Mẫu số không được bằng 0")
        self.tử = tu
        self.mẫu = mau
        self.rut_gon()

    def hop_le(self):
        return self.mẫu != 0

    def rut_gon(self):
        ucln = gcd(int(self.tử), int(self.mẫu))
        self.tu //= ucln
        self.mau //= ucln

    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        if self.mẫu == 0:
            raise ValueError("Mẫu số phải khác 0")
        self.rut_gon()

    def hien_thi(self):
        print(f"{self.tử}/{self.mẫu}")
```
