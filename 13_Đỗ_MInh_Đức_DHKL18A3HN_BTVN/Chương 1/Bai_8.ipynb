```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def __str__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"
```

```python
class Employee:
    def __init__(self, name, birth_date, join_date):
        self.name = name
        self.birth_date = birth_date
        self.join_date = join_date

    def __str__(self):
        return (f"Họ tên: {self.name}\n"
                f"Ngày sinh: {self.birth_date}\n"
                f"Ngày vào công ty: {self.join_date}")
```

```python
def nhap_date(thong_bao):
    print(thong_bao)
    d = int(input("  Ngày: "))
    m = int(input("  Tháng: "))
    y = int(input("  Năm: "))
    return Date(d, m, y)


def main():
    ds = []
    n = int(input("Nhập số nhân viên: "))

    for i in range(n):
        print(f"\nNhân viên {i+1}")
        name = input("Họ tên: ")

        birth = nhap_date("Nhập ngày sinh")
        join = nhap_date("Nhập ngày vào công ty")

        nv = Employee(name, birth, join)
        ds.append(nv)

    print("\n--- DANH SÁCH NHÂN VIÊN ---")
    for nv in ds:
        print(nv)
        print("-" * 30)


main()
```
