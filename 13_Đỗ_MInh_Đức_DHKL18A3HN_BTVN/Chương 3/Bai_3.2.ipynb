```python
import numpy as np
arr = np.arange(13)
print("Array arr:", arr)
print("Kiểu dữ liệu:", arr.dtype)
print("Kích thước:", arr.shape)
arr_odd = arr[arr % 2 == 1]
arr_even = arr[arr % 2 == 0]
print("Các phần tử lẻ:", arr_odd)
print("Các phần tử chẵn:", arr_even)
arr_update_1 = np.where(arr % 2 == 0, arr, 100)
print("Array cập nhật:", arr_update_1)
```
