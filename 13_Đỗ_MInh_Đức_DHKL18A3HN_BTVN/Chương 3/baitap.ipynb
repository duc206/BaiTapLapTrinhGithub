```python
import numpy as np
data = np.genfromtxt('drinks.csv', delimiter=',', dtype=None, names=True, encoding='utf-8')
print("Kiểu dữ liệu:", type(data))
print("Kích thước:", data.shape)
print("Các cột:", data.dtype.names)
print("5 dòng đầu:\n", data[:5])
print("5 dòng cuối:\n", data[-5:])
continents = np.unique(data['continent'])
for cont in continents:
    mean_beer = np.mean(data['beer_servings'][data['continent'] == cont])
    print(f"{cont}: trung bình beer_servings = {mean_beer:.2f}")
for cont in continents:
    wine_values = data['wine_servings'][data['continent'] == cont]
    print(f"{cont} - wine_servings: mean={np.mean(wine_values):.2f}, median={np.median(wine_values)}, max={np.max(wine_values)}, min={np.min(wine_values)}")
for cont in continents:
    subset = data[data['continent'] == cont]
    mean_beer = np.mean(subset['beer_servings'])
    mean_wine = np.mean(subset['wine_servings'])
    mean_spirit = np.mean(subset['spirit_servings'])
    print(f"{cont}: mean beer={mean_beer:.2f}, wine={mean_wine:.2f}, spirit={mean_spirit:.2f}")
for cont in continents:
    subset = data[data['continent'] == cont]
    print(f"{cont}: median beer={np.median(subset['beer_servings'])}, wine={np.median(subset['wine_servings'])}, spirit={np.median(subset['spirit_servings'])}")
sorted_data = np.sort(data, order='beer_servings')
print("5 quốc gia uống bia ít nhất:", sorted_data[:5]['country'])
print("5 quốc gia uống bia nhiều nhất:", sorted_data[-5:]['country'])
```

```python
import numpy as np
stocks1 = np.genfromtxt('stocks1.csv', delimiter=',', dtype=None, names=True, encoding='utf-8')
stocks2 = np.genfromtxt('stocks2.csv', delimiter=',', dtype=None, names=True, encoding='utf-8')
companies = np.genfromtxt('companies.csv', delimiter=',', dtype=None, names=True, encoding='utf-8', skip_header=1)
print("stocks1:", stocks1[:5])
print("stocks2:", stocks2[:5])
print("companies:", companies)
def fill_null(arr, col_name, func):
    col = arr[col_name].astype(float)
    isnan = np.isnan(col)
    for sym in np.unique(arr['symbol']):
        mask = (arr['symbol'] == sym)
        if col_name in ['high','low']:
            if col_name == 'high':
                val = np.nanmax(col[mask])
            else:
                val = np.nanmin(col[mask])
            col[mask & isnan] = val
    arr[col_name] = col
    return arr

stocks1 = fill_null(stocks1, 'high', np.nanmax)
stocks1 = fill_null(stocks1, 'low', np.nanmin)
stocks = np.concatenate((stocks1, stocks2))
stock_companies = []
for s in stocks:
    comp = [c for c in companies if c['name'] == s['symbol']]
    if comp:
        stock_companies.append(tuple(list(s) + list(comp[0])))
stock_companies = np.array(stock_companies, dtype=None)
for comp_name in np.unique([c[7] for c in stock_companies]):
    mask = np.array([c[7]==comp_name for c in stock_companies])
    open_mean = np.mean([float(c[2]) for c in stock_companies[mask]])
    close_mean = np.mean([float(c[5]) for c in stock_companies[mask]])
    volume_mean = np.mean([float(c[6]) for c in stock_companies[mask]])
    print(f"{comp_name}: mean open={open_mean}, close={close_mean}, volume={volume_mean}")
result = np.array(['up' if float(c[5])>float(c[2]) else 'down' for c in stock_companies])

```
