```python
import numpy as np
with open("Chương 3\height.txt", "r") as f:
    height = [float(x.strip()) for x in f.readlines() if x.strip() != ""]
with open("weights_1.txt", "r") as f:
    weight = [float(x.strip()) for x in f.readlines() if x.strip() != ""]
arr_height = np.array(height)
arr_weight = np.array(weight)
arr_height_m = arr_height * 0.0254
arr_weight_kg = arr_weight * 0.453592
arr_bmi = arr_weight_kg / (arr_height_m ** 2)
idx_example = min(50, len(arr_weight_kg)-1)
print(f"Cân nặng index {idx_example} (kg):", arr_weight_kg[idx_example])
start_idx, end_idx = 10, 20
arr_height_m_slice = arr_height_m[start_idx:end_idx+1]
print(f"Chiều cao index {start_idx}-{end_idx} (m):", arr_height_m_slice)
bmi_under_21 = arr_bmi[arr_bmi < 21]
print("Các cầu thủ có BMI < 21:", bmi_under_21)
print("Chiều cao trung bình (m):", np.mean(arr_height_m))
print("Cân nặng trung bình (kg):", np.mean(arr_weight_kg))
print("Chiều cao lớn nhất (m):", np.max(arr_height_m))
print("Cân nặng lớn nhất (kg):", np.max(arr_weight_kg))
print("Chiều cao nhỏ nhất (m):", np.min(arr_height_m))
print("Cân nặng nhỏ nhất (kg):", np.min(arr_weight_kg))
```
