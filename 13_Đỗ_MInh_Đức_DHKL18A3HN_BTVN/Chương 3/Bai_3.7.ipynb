```python
import numpy as np
np_baseball = np.genfromtxt("baseball_2D.txt", delimiter=",", dtype=float)
num_rows = np_baseball.shape[0]
np_height = np_baseball[:, 0]
np_weight = np_baseball[:, 1]
print("Dòng 50:", np_baseball[49])
print("Chiều cao cầu thủ 124:", np_height[123])
mean_height = np.mean(np_height)
mean_weight = np.mean(np_weight)
print("Chiều cao trung bình:", mean_height)
print("Cân nặng trung bình:", mean_weight)
correlation = np.corrcoef(np_height, np_weight)[0, 1]
print("Hệ số tương quan:", correlation)

```
